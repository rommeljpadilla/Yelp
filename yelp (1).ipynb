```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os, json, math, time
from yelpapi import YelpAPI
from tqdm.notebook import tqdm_notebook

```

```python
with open('C:/Users/RJ/.secret/yelp_api.json') as f:
    login = json.load(f)
login.keys()
```

```python
yelp = YelpAPI(login['api-key'], timeout_s=5)
yelp
```

```python
LOCATION = "Pasadena, CA, 91103"
TERM = 'Pizza'
```

```python
FOLDER = "Data/"
os.makedirs(FOLDER, exist_ok=True)
```

```python
JSON_FILE = f"{FOLDER}{TERM}_{LOCATION.split(',')[0]}.json"
JSON_FILE
```

```python
file_exists = os.path.isfile(JSON_FILE)
if file_exists == False:
    folder = os.path.dirname(JSON_FILE)
    
    if len(folder)>0:
        os.makedirs(folder,exist_ok=True)
        
    print("not found")
    
    with open(JSON_FILE,'w') as f:
        json.dump([],f)
        
else:
    print("already exists")
```

```python
results = yelp.search_query(term=TERM, location=LOCATION)
results
```

```python
results['total']
```

```python
results_per_page=len(results['businesses'])
results_per_page
```

```python
n_pages = math.ceil(results['total'] / results_per_page)
n_pages
```

```python
for i in tqdm_notebook( range(0, n_pages+1)):
    try:
        time.sleep(.2)
        with open(JSON_FILE, 'r') as f:
            previous_results = json.load(f)
            
        n_results = len(previous_results)
        
        results = yelp.search_query(term=TERM, location=LOCATION, offset=n_results+1)
        
        previous_results.extend( results['businesses'])
        
        with open(JSON_FILE,'w') as f:
            json.dump(previous_results, f)
            
    except Exception as e:
        print(e)
        break
```
